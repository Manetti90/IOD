# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
num_bins = 50
n, bins, patches = plt.hist(df.age, num_bins, facecolor='blue', alpha=0.5)

#labels & Titles
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
````

#### 4.B BoxPlot of `age`

````python
plt.boxplot(df.age)
plt.yscale('linear')
plt.grid(which='major', color='b', linestyle='-')
plt.grid(which='minor', color='r', linestyle='--')

#Labels & Titles

plt.title("Box Plot of of Age")
plt.ylabel("Age")


plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
plt.scatter(df["age"], df["length_of_service"])

plt.title("Scatterplot of of Age & Length of Service")
plt.xlabel("Age")
plt.ylabel("Length of Service")
plt.show()
````

#### 4.D Print a statistical summary of `age`

````python
print('statistical summary of observations =',stats.describe(df.age))
````
